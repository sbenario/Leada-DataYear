##Problem 2:
Which airport, as defined by the IATA code, has at least 10,000 flights and had the lowest probability for a delayed flight in the data?


_Define a delay as:_

	If ArrDelay > 0, then count as a delayed flight to Origin airport.  
	If DepDelay > 0, then count as a delayed flight to Dest airport.  
	Total delayed flights equals the sum of the two above.  
	Probability of delay equals total delayed flights divided by total flights (Origin & Dest) that go through that airport.  

```python
# Step 1 is filter our set down to airports that have 
# had at least 10,000 flights.

print df['Origin'].describe()
print
print df['Dest'].describe()
```

```python
dfOrigin = df.groupby('Origin', sort=False)  #faster if we don't sort. ...probably
dfDest = df.groupby('Dest', sort=False)

print dfOrigin.describe()
print dfOrigin.count()
```

```python
print "This is a pandas series"
print "This represents total entries for a given airport -- including cancelled flights, but I think that's fine"
print dfOrigin.size()   #I think "size" is the thing we want here... I basically want a count of the rows in each group


# If "size" weren't correct, here would be another way to get the same output:

#dfOrigin.indicies is a dictionary. Keys are Airport codes, Values is an array of row numbers for each entry.
#for (k, v) in dfOrigin.indices.items():
#    print (k, len(v))
```

```python
OriginFlightNums = dfOrigin.size()
DestFlightNums = dfDest.size()

totalFlights = {}

#we're going to do this with 2 quick iterations to create a dictionary, totalFlights, of total number of flights
for k, v in OriginFlightNums.iteritems():
    totalFlights[k] = v

for k,v in DestFlightNums.iteritems():
    if totalFlights.has_key(k):
        totalFlights[k] += v
    else:
        totalFlights[k] = v

#print totalFlights
print len(totalFlights)
```

```python
#lets prune everything under 10,000
totalFlights = {k: v for (k,v) in totalFlights.iteritems() if v >10000 }
print len(totalFlights)
#print totalFlights

print "totalFlights is now a dictionary of all airports with >10000 flights in the year."
```
