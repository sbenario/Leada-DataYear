##Problem 1:
Find the IATA code that appears in Origins, but not in Destinations.

```python
origins = pd.unique(df['Origin'])
dests = pd.unique(df['Dest'])
```

```python
# I'm pretty sure that some of the numpy operations we're going to use
# expect a sorted list. Either way, can't really hurt, since it doesn't 
# take long.

origins.sort()
dests.sort()
```

```python
# naive method of iterating through the dest and origin lists
# looking for the one that's missing.
for each in origins:
    if not each in dests:
        print each
```

```python
# Fast method using Numpy
print np.setdiff1d(origins,dests)
```

##Problem 2:
Which airport, as defined by the IATA code, has at least 10,000 flights and had the lowest probability for a delayed flight in the data?


_Define a delay as:_

	If ArrDelay > 0, then count as a delayed flight to Origin airport.  
	If DepDelay > 0, then count as a delayed flight to Dest airport.  
	Total delayed flights equals the sum of the two above.  
	Probability of delay equals total delayed flights divided by total flights (Origin & Dest) that go through that airport.  

```python
# Step 1 is filter our set down to airports that have 
# had at least 10,000 flights.

print df['Origin'].describe()
print
print df['Dest'].describe()
```

```python

dfOrigin = df.groupby('Origin', sort=False)  #faster if we don't sort%%!
dfDest = df.groupby('Dest', sort=False)
```

```python
dfOrigin.describe()
dfOrigin.count()
```

```python
# In the dataframe that's grouped by Origin, we want to count
# instances. The column we select is irrelevant.

countsOfOrigin = dfOrigin['Year'].count()
countsOfDest = dfDest['Year'].count()
print countsOfOrigin
```

```python
print type(countsOfOrigin)
#print countsOfOrigin
print countsOfDest['TUP']
print countsOfOrigin['TUP']
```

```python
originsDict = {k: countsOfOrigin[k] for k in countsOfOrigin.keys()}
flightsDict = {k: originsDict.get(k, 0) + countsOfDest[k] for k in countsOfDest.keys() if originsDict.get(k, 0) + countsOfDest[k] > 10000}
for k in originsDict.keys():
    if originsDict[k] > 10000 and k not in flightsDict:
        flightsDict[k] = originsDict[k]
print flightsDict

print len(flightsDict)
print len(originsDict)
```
